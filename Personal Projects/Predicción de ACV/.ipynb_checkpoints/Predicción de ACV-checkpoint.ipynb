# Predicción de ACV

El objetivo de este proyecto es encontrar patrones de comportamientos identificables que predisponen a una persona a sufrir ACV y así generar un modelo que prediga la probabilidad de una persona a sufrir un ACV.

Los datos fueron obtenidos de un centro de salud. El dataset cuenta de 11 caracteristicas clinicas que potencialmente podrian ser usadas en un modelo de predicción de ACV.

| Variable       |        Significado     | 
|-----------------------|-----------------------|
| id            |        Identificador      |
| gender         |        Género      |
| age            |        Edad     |
| hypertension   |        Si el paciente sufre de hipertensión      |
| heart_disease  |        Si el paciente sufre de efermedad cardiovascular|
| ever_married   |        Si estuvo o no casado/a     |
| work_type      |        Tipo de empleo    |
| Residence_type |        Si vive en zona rural o urbana    |
|avg_glucose_level |      Nivel medio de azúcar en sangre      |
| bmi              |      Indice de masa corporal    |
| smoking_status   |      Si es fumador, ex-fumador, nunca fumo o no se tienen datos    |
| stroke           |      1 si el paciente sufrio ACV(ACV+) y 0 si no sufrio ACV(ACV-)    |

```python
# importando librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, jaccard_score, log_loss, confusion_matrix, roc_curve, roc_auc_score, classification_report
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
import warnings

warnings.filterwarnings("ignore")
sns.set()
%matplotlib inline
```

```python
# importando el dataset
df = pd.read_csv("stroke_dataset.csv")
df.head()
```

# Análisis Exploratorio de datos (EDA)

Se realiza un análisis exploratorio de datos (EDA), que consiste en trazar gráficas para generar un análisis sobre el dataset. En esta instancia se buscan patrones, correlación entre variables y el entendimiento general del dataset.

#### Análisis categórico

```python
# obteniendo los nombres de las columnas object (columnas categoricas)
cat_vars = df.select_dtypes(include="object").columns.tolist()
exclude_vars = ["stroke"]
cat_vars = [var for var in cat_vars if var not in exclude_vars]

# creando una figura con los subplots
num_cols = len(cat_vars)
num_rows = (num_cols + 2) // 3
fig, axs = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, 5*num_rows))
axs = axs.flatten()

# creando grafico de barras para cada variable categorica
for i, var in enumerate(cat_vars):
    top_values = df[var].value_counts().nlargest(6).index
    filtered_df = df[df[var].isin(top_values)]
    ax = sns.countplot(x=var, data=filtered_df, ax=axs[i], hue="stroke")
    axs[i].set_title(var)
    axs[i].set_xlabel("")
    axs[i].tick_params(axis="x", rotation=90)

    # generando etiqueta de datos
    for valor in ax.patches:
        ax.annotate(f"{int(valor.get_height())}", 
                (valor.get_x() + valor.get_width() / 2., valor.get_height()),
                ha="center", va="center", xytext=(0, 10), textcoords="offset points")

# remueve los subplots que queden vacios
if num_cols < len(axs):
    for i in range(num_cols, len(axs)):
        fig.delaxes(axs[i])

# ajustando los espacios entre subplots
fig.tight_layout()
```

Tras realizar el análisis de los datos categóricos con respecto a la variable objetivo, podemos deducir lo siguiente:
- Género: Se evidencia un pequeño aumento en la probabilidad del desarrollo de un ACV en el género masculino, ya que el género masculino presentó una probabilidad de 5,38% de presentar un ACV mientras que el género femenino presento una probabilidad de 4,94%.
- Estado civil: El haber estado casado alguna vez puede tener relevancia a la hora de desarrollarse un ACV, se puede deducir que esto es debido a las nuevas responsabilidades que se adoptan y que pueden acarrear a un aumento del estrés.
- Tipo de trabajo: Se evidencia que los trabajos en términos proporcionales, independientemente de su naturaleza ya sea de tipo privado, autoempleado o empleado del gobierno tienen probabilidad de influir en el desarrollo del ACV.
- Tipo de residencia: El vivir en un entorno urbano debido a la naturaleza del ambiente, los hábitos y demás factores influyentes, existe una probabilidad un poco más alta de que una persona presente ACV+ con respecto a una persona que viva en un entorno rural.
- Fumador activo: El ser fumador o una persona que fumó anteriormente aumenta las probabilidades de ser un ACV+.

# Train Test Split

```python
df["stroke"].value_counts()
```

Debido a que existe un desbalanceo entre los datos objetivos, se deberá realizar un balanceo.

```python
X = df.drop(columns="stroke")    
y = df["stroke"]
```

```python
g = sns.countplot(data=df, x="stroke", palette="pastel")
g.set(title="Cantidad de pacientes ACV+/ACV-", ylabel ="Número de pacientes", xlabel="")
g.set_xticklabels(["ACV-", "ACV+"])
for valor in g.patches:
        g.annotate(f"{int(valor.get_height())}", 
                (valor.get_x() + valor.get_width() / 2., valor.get_height()),
                ha="center", va="center", xytext=(0, 10), textcoords="offset points")
plt.show()
```

Se usará SMOTE, este generará muestras sintéticas para la clase minoritaria con el objetivo de equilibrar la distribución de clases. Creando nuevos ejemplos sintéticos mediante la interpolación de características de ejemplos vecinos.

```python
smote = SMOTE(random_state=2)
X, y = smote.fit_resample(X, y)
```

```python
g = sns.countplot(x=y, palette="pastel")
g.set(title="Cantidad de pacientes ACV+/ACV-", ylabel ="Número de pacientes", xlabel="")
g.set_xticklabels(["ACV-", "ACV+"])
for valor in g.patches:
        g.annotate(f"{int(valor.get_height())}", 
                (valor.get_x() + valor.get_width() / 2., valor.get_height()),
                ha="center", va="center", xytext=(0, 10), textcoords="offset points")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2, shuffle=True)
```

# Construcción del modelo

Usaré los siguientes modelos de aprendizaje automático para predecir si una persona tiene la probabilidad de ser un ACV+:

- Regresión logística: Se utiliza para predecir una variable categórica binaria (dos categorías) en función de una o más variables predictoras.
- K-Nearest Neighbors (k-NN): Utiliza una función de distancia para encontrar los k observaciones más cercanas a un nuevo caso y clasificarlo en función de la mayoría de las categorías de los k vecinos más cercanos.
- Random Forest: Es un modelo que combina múltiples árboles de decisión entrenados en diferentes subconjuntos de datos y características.

# Regresión logística

```python
log_reg = LogisticRegression(max_iter=1000)
```

```python
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(X_train, y_train)
```

# Evaluación del modelo


La **precisión** es la proporción de predicciones positivas correctas (verdaderos positivos) respecto a todas las predicciones positivas hechas por el modelo (verdaderos positivos + falsos positivos). Mide la capacidad del modelo para no etiquetar incorrectamente una muestra como positiva.

El **recall** se define como la proporción de verdaderos positivos (TP) que se identifican correctamente entre todos los verdaderos positivos y falsos negativos (FN) en el conjunto de datos. Mide la capacidad del modelo para encontrar todas las muestras positivas. 

El **F1 score** es una métrica comúnmente utilizada para evaluar la precisión de un modelo de clasificación. Es la media armónica entre la precisión y el recall del modelo.

```python
log_reg_pred = log_reg.predict(X_test)
```

```python
print(classification_report(y_pred=log_reg_pred, y_true=y_test))
```

#### Matriz de confusión

La matriz de confusión permite visualizar cuántas instancias se clasificaron correctamente (verdaderos positivos y verdaderos negativos) y cuántas se clasificaron incorrectamente (falsos positivos y falsos negativos). En este caso, es de vital importancia  minimizar los falsos negativos debido a que al ser un modelo de predicción de un enfermedad potencialmente mortal o  fisicamente inhabilitante, informar un falso negativo a un paciente puede evitar que reciba el tratamiento adecuado a tiempo.

```python
# evaluación del modelo de regresión logística
log = sns.heatmap(confusion_matrix(y_test, log_reg_pred), annot=True, fmt="d", cmap='Blues')
log.set(ylabel ="Valor real", xlabel="Predicción")
log.set_xticklabels(["ACV-", "ACV+"]) 
log.set_yticklabels(["ACV-", "ACV+"])
title = " Precisión de la Regresión Logística: {0}".format(accuracy_score(y_test, log_reg_pred).round(2))
plt.title(title, size = 15)

confusion_matrix(y_test, log_reg_pred)
```

# k-Nearest Neightbors (k-NN)

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_train, y_train)
```

# Evaluación del modelo

```python
knn_pred = knn.predict(X_test)
```

```python
print(classification_report(y_pred=knn_pred, y_true=y_test))
```

#### Matriz de confusión

```python
# evaluación del modelo de k-NN
kn = sns.heatmap(confusion_matrix(y_test, knn_pred), annot=True, fmt="d", cmap='Blues')
kn.set(ylabel ="Valor real", xlabel="Predicción")
kn.set_xticklabels(["ACV-", "ACV+"]) 
kn.set_yticklabels(["ACV-", "ACV+"])
title = " Precisión del k-NN: {0}".format(accuracy_score(y_test, knn_pred).round(2))
plt.title(title, size = 15)

confusion_matrix(y_test, knn_pred)
```

# Random Forest Classifier

```python
rfc = RandomForestClassifier()
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 5, 10],
    "max_features": ["sqrt", "log2", None],
    "random_state": [0, 42]
}

# desarrollando un grid search con validación cruzada para encontrar los mejores hiperparametros
grid_search = GridSearchCV(rfc, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Print the best hyperparameters
print(grid_search.best_params_)
```

```python
rfc = RandomForestClassifier(random_state=42, max_depth=None, max_features="sqrt", n_estimators=200)
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_train, y_train)
```

# Evaluación del modelo

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_pred=rfc_pred, y_true=y_test))
```

#### Matriz de confusión

```python
# evaluación del modelo de k-NN
rf = sns.heatmap(confusion_matrix(y_test, rfc_pred), annot=True, fmt="d", cmap='Blues')
rf.set(ylabel ="Valor real", xlabel="Predicción")
rf.set_xticklabels(["ACV-", "ACV+"]) 
rf.set_yticklabels(["ACV-", "ACV+"])
title = " Precisión del Random Forest: {0}".format(accuracy_score(y_test, rfc_pred).round(2))
plt.title(title, size = 15)

confusion_matrix(y_test, rfc_pred)
```

Luego de un detallado análisis de las variables del dataset y de como esta influyen en los grupos ACV+ y ACV-, se procedió a la prueba de diferentes modelos de predicción a partir del dataset. A continuación, se resumen las metricas de los modelos evaluados:


| Modelo                     | Precision             | Recall                | F1-score              | Falsos negativos| 
|----------------------------|-----------------------|-----------------------|-----------------------|-----------------|

El algoritmo de k-NN se considera el más adecuado para realizar la predicción. Si bien no predice con la misma precisión que un modelo de Random Forest, el k-NN generó una menor cantidad de falsos negativos que el resto de los modelos (es decir pacientes posiblemente ACV+ que son etiquetados como ACV-). En este caso, al tratarse de predicción relacionada a una enfermedad, es de vital importancia que este número sea lo más bajo posible.

```python
def prediccion_acv (data):
    input_data_as_numpy_array= np.asarray(data)

    input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

    prediction = knn.predict(input_data_reshaped)
    
    print(f'El resultado es {prediction[0]}')

    if (prediction[0]== 0):
      print("La persona es ACV-")
    else:
      print("La persona es ACV+")
```

```python
datos_prediccion = (0,80,1,0,1,2,1,83.75,28.89,2) # persona ACV-
prediccion_acv(datos_prediccion)
```
