# Predicción de precio de casas

Este proyecto se desarrolla para predecir el precio de casas en función de la variables como el número de habitaciones, tamaño de la casa, ubicación, entre otros.

```python
# importando librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
import folium
from sklearn import metrics
from folium.plugins import FastMarkerCluster
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
import warnings

warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

# Importando el dataset

```python
df = pd.read_csv("home_data.csv")
df.head()
```

# Procesamiento del dataset

```python
# eliminando columnas innecesarias
df.drop(columns=["id", "date"], inplace=True)
```

```python
# buscando valores Null
df.isnull().sum()
```

```python
# cambiando tipo de dato
df["bathrooms"] = df["bathrooms"].astype(int)
df["floors"] = df["floors"].astype(int)
```

```python
# renombrando la columna yr_built a age y cambiando sus datos
df.rename(columns={"yr_built":"age"},inplace=True)
df["age"] = 2023 - df["age"]

# renombrando la columna yr_renovated a renovated y cambiando sus valores a 0 y 1
df.rename(columns={"yr_renovated":"renovated"},inplace=True)
df["renovated"] = df["renovated"].apply(lambda x: 0 if x == 0 else 1)
df.head()
```

# Análisis del dataset

```python
print(f"Este dataset contiene {df.shape[0]} filas y {df.shape[1]} columnas")
```

# Train/Test Split

```python
X = df.drop("price",axis=1)
y = df["price"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Construcción del modelo

Usaré los siguientes modelos de aprendizaje automático para predecir el precio de las casas:

- Regresión lineal polinomial.
- Ridge Regression.
- Random Forest Regression.

# Regresión lineal polinomial

```python
input = [("scale", StandardScaler()), ("polynomial", PolynomialFeatures(degree=2)) ,("model", LinearRegression())]
rlp = Pipeline(input)
```

```python
rlp.fit(X_train, y_train)
```

```python
rlp.score(X_train, y_train)
```

```python
rlp_pred = rlp.predict(X_test)
r2_score(y_test, rlp_pred)
```

# Regresión Ridge

```python
rr = Ridge(alpha = 0.001)
```

```python
rr.fit(X_train, y_train)
```

```python
rr.score(X_train, y_train)
```

```python
rr_pred = rr.predict(X_test)
r2_score(y_test, rr_pred)
```

# Random Forest Regression

```python
rfr = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_train, y_train)
```

```python
rfr_pred = rfr.predict(X_test)
r2_score(y_test, rfr_pred)
```

# Evaluación de los modelos

#### Distribución

```python
fig, ax = plt.subplots(1,3,figsize=(20,5))
# Regresión lineal
sns.distplot(y_test,ax=ax[0])
sns.distplot(rlp_pred,ax=ax[0])
# Ridge Regression
sns.distplot(y_test,ax=ax[1])
sns.distplot(rr_pred,ax=ax[1])
# Random Forest Regression
sns.distplot(y_test,ax=ax[2])
sns.distplot(rfr_pred,ax=ax[2])

# leyendas
ax[0].legend(["Precio actual", "Predicción"])
ax[1].legend(["Precio actual", "Predicción"])
ax[2].legend(["Precio actual", "Predicción"])

# titulos
ax[0].set_title("Regresión Lineal")
ax[1].set_title("Ridge Regression")
ax[2].set_title("Random Forest Regression")
plt.show()
```

#### Errores

```python
fig, ax = plt.subplots(1,3,figsize=(20,5))
sns.barplot(x=["Regresión Lineal","Ridge Regression","Random Forest"], y=[mean_absolute_error(y_test, rlp_pred), mean_absolute_error(y_test, rr_pred), mean_absolute_error(y_test, rfr_pred)], ax=ax[0])
sns.barplot(x=["Regresión Lineal","Ridge Regression","Random Forest"], y=[mean_squared_error(y_test, rlp_pred), mean_squared_error(y_test, rr_pred), mean_squared_error(y_test, rfr_pred)], ax=ax[1])
sns.barplot(x=["Regresión Lineal","Ridge Regression","Random Forest"], y=[np.sqrt(mean_squared_error(y_test, rlp_pred)), np.sqrt(mean_squared_error(y_test, rr_pred)), np.sqrt(mean_squared_error(y_test, rfr_pred))], ax=ax[2])

ax[0].set_ylabel("Mean Absolute Error")
ax[1].set_ylabel("Mean Squared Error")
ax[2].set_ylabel("Root Mean Squared Error")
plt.show()
```

#### Accuracy

```python
models = ["Linear Regression", "Ridge Regression", "Random Forest Regression"]
accuracy = [metrics.r2_score(y_test, rlp_pred),metrics.r2_score(y_test, rr_pred),metrics.r2_score(y_test, rfr_pred)]
plt.figure(figsize=(8,6))
sns.barplot(x=models, y=accuracy)

for i, val in enumerate(accuracy):
    plt.text(i, val, f'{val:.4f}', ha="center", va="bottom", fontsize=12)
    
plt.title("Comparación de Accuracy")
plt.ylabel("Accuracy")
plt.show()
```

# Conclusión

Del análisis, podemos ver que el modelo Random Forest funcionó mejor que el modelo Ridge Regression y el modelo Regresión Lineal, ya que tuvo una mejor precisión y un menor valor de promedio de errores.

Durante el proceso de EDA, descubrimos que la ubicación de la casa es un factor muy importante para determinar el precio de la casa, ya que casas con área similar y otras características pueden tener precios diferentes dependiendo de la ubicación de la casa.

La ubicación de las casas se ha trazado en el mapa utilizando los valores de longitud y latitud, lo que aclara el papel de la ubicación a la hora de determinar el precio de la casa.
