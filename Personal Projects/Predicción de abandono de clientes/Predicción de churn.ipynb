# **1. Introduccin**

En la industria de las telecomunicaciones, los clientes pueden elegir entre una variedad de empresas para satisfacer sus necesidades de comunicacin e internet. Los clientes son muy crticos con el tipo de servicios que reciben y juzgan toda la empresa en funcin de una sola experiencia. 

Estos servicios de comunicacin se han vuelto tan recurrentes e inseparables de la rutina diaria que una pausa de mantenimiento de 30 minutos genera ansiedad en los usuarios, resaltando nuestra actitud de dar por sentado estos servicios! Sumado a los altos costos de adquisicin de clientes, el anlisis de la rotacin se vuelve muy crucial. La tasa de rotacin es una mtrica que describe el nmero de clientes que cancelaron o no renovaron su suscripcin con la empresa. 

Por lo tanto, cuanto mayor sea la tasa de rotacin, ms clientes dejarn de comprar en su negocio, afectando directamente los ingresos. Por lo tanto, basndose en los conocimientos adquiridos del anlisis de la rotacin, las empresas pueden construir estrategias, segmentar objetivos, mejorar la calidad de los servicios proporcionados para mejorar la experiencia del cliente, cultivando as la confianza con los clientes. Por eso construir modelos predictivos y crear informes de anlisis de rotacin se vuelve fundamental que allana el camino para el crecimiento!

## **1.2 Que es el *Churn*?**


La prediccin del abandono de clientes implica identificar a los clientes en riesgo que probablemente cancelarn sus suscripciones o cerrarn/abandonarn sus cuentas. Un modelo de rotacin funciona pasando datos de clientes anteriores a travs de un modelo de aprendizaje automtico para identificar las conexiones entre las caractersticas y los objetivos, y hacer predicciones sobre nuevos clientes.

La ventaja de calcular la tasa de abandono de una empresa es que proporciona claridad sobre qu tan bien la empresa est reteniendo a los clientes, lo que es un reflejo de la calidad del servicio que brinda la empresa, as como de su utilidad.

Si una empresa ve que su tasa de abandono aumenta de un perodo a otro, entonces comprende que un componente fundamental de la forma en que administra su negocio es defectuoso. La empresa puede estar ofreciendo un producto defectuoso, puede tener un mal servicio al cliente o su producto puede no ser atractivo para las personas que decidieron que el costo no vale la pena .

La tasa de abandono le indicar a una empresa que necesita comprender por qu sus clientes se van y dnde arreglar su negocio. El costo de adquirir nuevos clientes es mucho mayor que el de retener a los clientes actuales, por lo que, al asegurarse de que los clientes por los que trabaj duro para atraer sigan siendo clientes de pago, tiene sentido comprender la calidad de su negocio.


## **1.3 Como se calcula el *Churn Rate*?**



Digamos que tenas 1000 clientes al comienzo del mes, y para el final del mismo, solo tienes 800 de esos 1000 restantes. Eso significa que perdiste 200 clientes durante ese perodo, as que divides 200 entre 1000 y luego multiplicas por 100. Tu resultado es una tasa de rotacin del 20%, lo cual es bastante alto.

Dependiendo de tu modelo de negocio, debes determinar el marco de tiempo adecuado para el cual deseas calcular la tasa de rotacin, como mensual, trimestral, anual o de todos los tiempos.

## **1.4 Importancia de predecir el *Churn Rate* de clientes**

**Como empresa, ser capaz de predecir el churn es el primer paso para evitarlo. Si eres capaz de predecir el abandono de clientes, tambin podrs prevenirlo con las medidas adecuadas.**

La salud financiera de una empresa est relacionada con la tasa de abandono. Para seguir siendo competitivo en el mercado y crecer, la cantidad de nuevos consumidores cada mes debe ser mayor que la tasa de abandono.

Si el nmero de cancelaciones es mayor que el nmero de suscripciones, por ejemplo, es seal de que algo anda mal en la operacin. Por lo tanto, es fundamental estar atento a este ritmo, para poder identificar tempranamente cualquier problema en el desempeo del negocio.

**Estas son algunas de las razones por las que debes prevenir el churn rate de clientes:**

1. La adquisicin de nuevos clientes es costosa y supone un gran esfuerzo, mientras que la prevencin del abandono de clientes es mucho ms rentable. 

2. La competencia es cada vez mayor, los productos son a menudo similares o incluso intercambiables. Por lo tanto, la probabilidad de abandono es cada vez mayor. Por lo tanto, toda empresa debe tener estrategias para la retencin de clientes a largo plazo. 

3. Hay muchos factores diferentes que conducen al churn rate de clientes y es importante que las empresas identifiquen, comprendan y eliminen cada uno de ellos para que los clientes sigan vinculados a la empresa. 

4. Identificar estos factores de prdida de clientes es relativamente fcil con las herramientas adecuadas de gestin de la experiencia del cliente. Las encuestas peridicas y los anlisis de los puntos de contacto a lo largo de todo el recorrido del cliente son el paso ms importante en este sentido.

### **Tcnicas de Remuestreo**

* **SMOTE (Synthetic Minority Over-sampling Technique):**

SMOTE es una tcnica de sobremuestreo utilizada para abordar el desequilibrio de clases en conjuntos de datos, especficamente cuando la clase minoritaria est subrepresentada. Lo que hace SMOTE es generar muestras sintticas para la clase minoritaria mediante la interpolacin de instancias existentes en el espacio de caractersticas. Bsicamente, SMOTE crea nuevas instancias sintticas de la clase minoritaria mediante la combinacin de ejemplos cercanos de la misma clase.

* **RandomUnderSampler**:

El RandomUnderSampler es una tcnica de remuestreo utilizada en el contexto de problemas de desequilibrio de clases en conjuntos de datos de aprendizaje automtico.

El RandomUnderSampler aborda este problema seleccionando aleatoriamente una muestra de la clase mayoritaria (o clases) para que est ms equilibrada con respecto a la clase minoritaria. Esto se logra eliminando aleatoriamente instancias de la clase mayoritaria hasta que la proporcin de clases en el conjunto de datos sea ms equitativa.

```python
#Definimos las funciones para el remuestreo

over = ADASYN(sampling_strategy=0.5)
under = RandomUnderSampler(sampling_strategy= 0.5)
```
