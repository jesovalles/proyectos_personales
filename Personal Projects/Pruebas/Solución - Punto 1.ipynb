# Solución del Punto 1 - Carga de información

```python
# Importar librerías
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

# Transformación y carga del dataset

```python
df = pd.read_csv('OFEI1204.txt', sep='\t', header=None) # Leer el archivo de texto

df = df.dropna() # Eliminar filas vacías

df = df[0].str.split(',') # Dividir las columnas

df= df.apply(pd.Series) # Transformar a dataframe

# Crear columna Agente
df['Agente'] = df.iloc[:, 0].apply(lambda x: x.split('AGENTE: ')[-1] if x.startswith('AGENTE:') else np.nan) 

df.columns = ["Planta", "Registro"] + [f'Hora_{i+1}' for i in range(24)] + ["Agente"] # Renombrar columnas

df = df.reindex(columns=['Agente', 'Planta', 'Registro', 'Hora_1', 'Hora_2', 'Hora_3', 'Hora_4', 'Hora_5', 
                         'Hora_6', 'Hora_7', 'Hora_8', 'Hora_9', 'Hora_10', 'Hora_11', 'Hora_12', 'Hora_13', 
                         'Hora_14', 'Hora_15', 'Hora_16', 'Hora_17', 'Hora_18', 'Hora_19', 'Hora_20', 
                         'Hora_21', 'Hora_22', 'Hora_23', 'Hora_24']) # Reordenar columnas

df['Agente'] = df['Agente'].ffill() # Rellenar columna Agente

df = df[~df['Planta'].str.startswith('AGENTE:')] # Eliminar filas innecesarias

df = df[df['Registro'] == ' D'] # Filtrar solo por Registro 'D'

df.drop(columns=['Registro'], inplace=True) # Eliminar columna Registro

df.reset_index(drop=True, inplace=True) # Restablecer el índice

df.to_csv('OFEI1204_NEW.csv', index=False) # Guardar archivo transformado

df.head(10) # Visualizar las primeras 10 filas
```

Este dataset que contiene solo tipo de registro D, está conformado por 305 filas y 26 columnas.
