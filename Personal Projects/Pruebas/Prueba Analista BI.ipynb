## Resolución de prueba - Analista BI

La resolución de esta prueba implica explorar los datos de transacciones de clientes, identificar patrones y tendencias importantes, y luego aplicar técnicas de segmentación para agrupar clientes similares con el objetivo de obtener información valiosa para la toma de decisiones empresariales.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from datetime import datetime

warnings.filterwarnings('ignore')
```

### Visualización del dataset

```python
df = pd.read_csv("data_proceso_Analista_BI.csv")
df.head()
```

Este dataset contiene 9955898 filas y 5 columnas.

### Limpieza y transformación de los datos

Antes de realizar cualquier tipo de análisis debemos chequear la calidad de los datos para corregir inconsistencias y no crear sesgos que nos puedan afectar en nuestro análisis.

```python
df.isnull().sum()
```

Acá podemos darnos cuenta que existen valores faltantes en las columnas "codigo_cliente" y "valor de venta". Podemos suponer que los codigo_cliente faltantes se deben a que el cliente no se quiso registrar y para los valor_venta faltantes, posiblemente se deban a algún error en la transacción. Además se encontraron ventas con valor de 0.0 las cuales no nos sirven para el análisis. Estas inconsistencias pueden alterar nuestro análisis, lo resolveremos creando la categoría "Desconocido" en la columna codigo_cliente y eliminando los registros vacios de la columna valor_venta así como las ventas con valor de 0.0. Adicional, se procede a cambiar el tipo de dato a las columnas codigo_producto y codigo_categoria a tipo string y la columna fecha_venta a tipo date/time para la posterior extracción de la fecha y hora que serán usadas para el análisis luego.

### Dataset transformado y limpio

```python
df["codigo_cliente"] = df["codigo_cliente"].fillna("Desconocido")

df = df.dropna(subset=["valor_venta"])
df = df.drop(df[df['valor_venta'] == 0.0].index)

df["codigo_producto"] = df["codigo_producto"].astype(str)
df["codigo_categoria"] = df["codigo_categoria"].astype(str)
df["fecha_venta"] = pd.to_datetime(df["fecha_venta"])

df['fecha'] = df['fecha_venta'].dt.date
df['hora'] = df['fecha_venta'].dt.time

df.drop(columns=["fecha_venta"], inplace=True)

df['hora'] = pd.to_datetime(df['hora'].apply(lambda x: x.strftime('%H:%M:%S')))

df['hora'] = df['hora'].dt.round('H')

df.head()
```

### Análisis exploratorio de datos

#### Top 5 productos que más ventas generaron

#### Registro de ventas

```python
# obteniendo los nombres de las columnas object (columnas categoricas)
cat_vars = df.select_dtypes(include='object').columns.tolist()

# creando una figura con los subplots
num_cols = len(cat_vars)
num_rows = (num_cols + 2) // 3
fig, axs = plt.subplots(nrows=num_rows, ncols=3, figsize=(15, 5*num_rows))
axs = axs.flatten()

# creando grafico de barras para cada variable categorica
for i, var in enumerate(cat_vars):
    top_values = df[var].value_counts().nlargest(5).index
    filtered_df = df[df[var].isin(top_values)]
    ax = sns.countplot(x=var, data=filtered_df, ax=axs[i])
    axs[i].tick_params(axis='x', rotation=90)
    axs[i].set_ylabel("Cantidad de registro de ventas")

# remueve los subplots que queden vacios
if num_cols < len(axs):
    for i in range(num_cols, len(axs)):
        fig.delaxes(axs[i])

# ajustando los espacios entre subplots
fig.tight_layout()
```

Primeramente realizamos un gráfico de barras mostrando un Top 5 de cantidad de ventas por cliente, producto, categoría, fecha y franja horaria. Conocer esta información es sumamente relevante ya que nos permite identificar cuales son las variables que más se venden. Hay un punto resaltante acá ya que se puede observar que el 46.43% de las ventas registradas corresponden al cliente ad2c9e0a9c. En el mismo orden de ideas se visualiza que el 74.47% de las compras se enfocan en la categoría 6 y el producto más vendido es el 33 que representa 27.28% de los registros totales. Esto nos da un panorama más amplio de donde enfocar más nuestros esfuerzos en lograr ventas. No hacemos mención de la fecha ya que podemos visualizar que maneja un promedio similar de registros de ventas.

#### Manejo de valores atípicos

Debido a los valores atípico que encontramos en la columna valor_venta, antes de realizar alguna gráfica de distribución debemos manejarlos ya que estos alterarán de gran manera la distribución de los datos. Para la correción de estos valores se usará el método de rango intercuartílico.

```python
Q1 = df['valor_venta'].quantile(0.25)
Q3 = df['valor_venta'].quantile(0.75)

IQR = Q3 - Q1 # Calcular el rango intercuartílico (IQR)

lower_limit = Q1 - 1.5 * IQR # Definir límites para detectar outliers
upper_limit = Q3 + 1.5 * IQR

df = df[(df['valor_venta'] >= lower_limit) & (df['valor_venta'] <= upper_limit)] # Filtrar valores que están dentro del rango límite
```

Luego de realizar el método de rango intercuartílico, podemos aún identificar valores atípicos pero no son tan significantes y podemos proceder a analizar nuestras gráficas. ### Segmentación de clientes

Esta segmentación de clientes se hará tomando en cuenta la cantidad de dinero gastado por cliente, agruparemos sus compras totales y en base a eso y a los quartíles estableceremos las categorías de la segmentación, categorizaremos a los clientes según una prioridad. Cabe destacar que como existen clientes con muchas ventas generadas, se realizará un proceso de rango intercuartílico para omitir esas particularidades y mantener nuestra distribución lo más real posible.

```python
clientes = df.groupby('codigo_cliente')['valor_venta'].sum().sort_values(ascending=False).reset_index()
```

```python
clientes.describe()
```

Luego de realizar un proceso de rango intercuartílico y con ayuda de la estadística de las ventas, se establecen los límites para la categorización usando las ventas totales de los cliente, estos límites son:
- Para cliente con prioridad baja será de [0, 499999)
- Para cliente con prioridad media será de [500000, 999999)
- Para cliente con prioridad alta será de [1000000, infinito)

```python
clientes
```

#### Categorías de clientes

```python
plt.figure(figsize=(12, 5))
sns.countplot(x='prioridad', data=clientes)
plt.ylabel("Número de clientes")
plt.show()
```

```python
Q11 = clientes['valor_venta'].quantile(0.25)
Q33 = clientes['valor_venta'].quantile(0.75)

IQRR = Q33 - Q11 # Calcular el rango intercuartílico (IQR)

lower_limit = Q11 - 1.5 * IQRR # Definir límites para detectar outliers
upper_limit = Q33 + 1.5 * IQRR

clientes = clientes[(clientes['valor_venta'] >= lower_limit) & (clientes['valor_venta'] <= upper_limit)] 
```

```python
Q11 = clientes['valor_venta'].quantile(0.25)
Q33 = clientes['valor_venta'].quantile(0.75)

IQRR = Q33 - Q11 # Calcular el rango intercuartílico (IQR)

lower_limit = Q11 - 1.5 * IQRR # Definir límites para detectar outliers
upper_limit = Q33 + 1.5 * IQRR

clientes = clientes[(clientes['valor_venta'] >= lower_limit) & (clientes['valor_venta'] <= upper_limit)] 
```

```python
# Definir los límites de los intervalos
bins = [0, 499999, 1499999, float('inf')]  # Los límites son [0, 499999), [500000, 999999), [1000000, infinito)

# Definir las etiquetas para las categorías
labels = ['Baja', 'Media', 'Alta']

# Crear la nueva columna categorizando los clientes
clientes['prioridad'] = pd.cut(clientes['valor_venta'], bins=bins, labels=labels, right=False)
```
