# Solución del Punto 2 - Manipulación de datos

```python
# Importar librerías
import pandas as pd
pd.set_option('display.max_columns', None)
```

# Carga del dataset 1

```python
df_1 = pd.read_excel('Datos Maestros VF.xlsx') # Cargar archivo excel

# Seleccionar columnas a usar
df_1 = df_1[['Nombre visible Agente', 'AGENTE (OFEI)', 'CENTRAL (dDEC, dSEGDES, dPRU…)',
             'Tipo de central (Hidro, Termo, Filo, Menor)']]

# Filtrar por agente EMGESA S.A. y Tipo de central H o T
df_1 = df_1[(df_1['AGENTE (OFEI)'] == 'EMGESA S.A.') &
        (df_1['Tipo de central (Hidro, Termo, Filo, Menor)'].isin(['H', 'T']))
        ]

df_1.head() # visualizar dataset
```

# Carga del dataset 2

```python
# Cargar archivo txt
df_2 = pd.read_csv('dDEC1204.txt', sep=',', header=None, encoding='ISO-8859-1') 

# Asignando nombre a la primera columna
df_2 = df_2.rename(columns={0: 'Central'}) 

# visualizar dataset
df_2.head() 
```

# Dataset final

```python
# Crear dataframe usando df_1 y df_2
df = pd.merge(df_1, df_2, left_on='CENTRAL (dDEC, dSEGDES, dPRU…)', right_on='Central', how='inner')

# Eliminar columna Central
df.drop(columns=['Central'], inplace=True)

# Crear columna que sume las horas horizontalmente de cada planta
df['total horas'] = df.select_dtypes(include='int').sum(axis=1)

# Filtrar por total horas mayores a 0
df = df[df['total horas'] > 0]

df.to_csv('df.csv', index=False) # Guardar archivo transformado

# Visualizar dataframe
df.head()
```
