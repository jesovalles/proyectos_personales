Primeramente se cargan las librerías a usar para el análisis.

```python
# importando librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

## Cargando el dataset

Procedemos a cargar el dataset en un dataframe.

```python
# cargando el dataset
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

Procedemos a revisar la estructura del dataframe, es decir, la cantidad de filas, columnas, tipos de datos, etc.

```python
# revisando informacion del dataframe
df.info()
```

## Procesamiento del dataframe

Para este análisis usaremos solo las columnas neighbourhood_group, neighbourhood, room_type, price, minimum_nights, reviews_per_month, ya que a criterio personal, son las que considero más relevantes.

```python
# seleccionando columnas para el análisis
df = df[['neighbourhood_group', 'neighbourhood', 'room_type', 'price', 'minimum_nights', 'reviews_per_month']]
df.head()
```

Anteriormente se pudieron visualizar valores nulos en la columna reviews_per_month, procedemos a chequear todas las columnas.

```python
# chequeando valores nulos
df.isnull().sum()
```

Al verificar que existen valores nulos en la columna reviews_per_month, procedemos a reemplazar los valores nulos por 0, esto debido a la naturaleza de la columna reviews_per_month, es decir, los valores nulos significan que las propiedades no tuvieron vistas, y por lo tanto no se generan sesgos en los datos.

```python
# reemplazando valores nulos por 0, ya que los nulos significan que no han tenido review
df['reviews_per_month'] = df['reviews_per_month'].replace(np.nan, 0)
df.head()
```

Se pudo evidenciar que existen registros de precios con valor de 0, estos registros no nos serán relevantes en nuestro análisis y por lo tanto procederemos a eliminarlos.

```python
# eliminando los registros donde el precio es 0
df = df[df['price'] != 0]
```

Crearemos una columna llamada price_night que nos mostrará lo que sería el precio por noche de las propiedades, será la división entre el precio y las noches mínimas por propiedad, ya que esto nos ayudará a tener un mejor contexto de los datos a nivel de precios.

```python
# creando columna price_night
df['price_night'] = df['price'] / df['minimum_nights']
df.head()
```

Para nuestro análisis tendremos un punto de partida donde el precio por noche tiene que ser mayor o igual a 10, esto para evitar valores anómalos, sin embargo de ser requerido, más adelante se podrá realizar algún método más avanzado para eliminarlos. Procedemos a filtrar el dataframe.

```python
# eliminando los registros donde el precio por noche sean menores a 10
df = df[df['price_night'] >= 10]
```

De esta manera ya tenemos nuestro dataframe preparado, por lo cual ya podemos realizar un análisis exploratorio de datos.

### Precio promedio y máximo por noche vs Zona

```python
# promedio del precio por noche
df_avg = df.groupby('neighbourhood_group')['price_night'].mean().sort_values(ascending=False).reset_index()
df_avg = df_avg.rename(columns={'neighbourhood_group': 'Zona', 'price_night': 'Precio promedio por noche'})

# máximo precio por noche
df_max = df.groupby('neighbourhood_group')['price_night'].max().sort_values(ascending=False).reset_index()
df_max = df_max.rename(columns={'neighbourhood_group': 'Zona', 'price_night': 'Precio máximo por noche'})

# configurando gráfico promedio por noche
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(df_avg['Zona'], df_avg['Precio promedio por noche'])
plt.xlabel('Zona')
plt.ylabel('Precio promedio por noche')
plt.title('Precio promedio por noche por Zona')
plt.xticks(rotation=45)

# configurando gráfico máximo precio por noche
plt.subplot(1, 2, 2)
plt.bar(df_max['Zona'], df_max['Precio máximo por noche'])
plt.xlabel('Zona')
plt.ylabel('Precio máximo por noche')
plt.title('Precio máximo por noche por Zona')
plt.xticks(rotation=45)

# ajustar diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
```

Manhattan es por largo, la zona donde el precio promedio por noche es mayor, superando en promedio los 100 y es razonable ese análisis ya que es la zona con más influencia tanto a nivel económico como social de New York. Sorpresivamente le sigue Staten Island con un precio promedio por noche rondando los 80. Ahora, lo sorprendente es lo máximo que se ha pagado por noche en dichas zonas, donde lo máximo que se pagó por noche fueron 8000. Estos valores posiblemente son atípicos, cuando se visualicen los vecindarios, podremos confirmar si dichos precios son en zonas de clase alta.

### Precio promedio por noche vs Vecindario

```python
# promediando el precio por noche
df_hood_avg = df.groupby(['neighbourhood', 'neighbourhood_group'])['price_night'].mean().sort_values(ascending=False).reset_index()
df_hood_avg = df_hood_avg.rename(columns={'neighbourhood': 'Vecindario', 
                                          'price_night': 'Precio promedio por noche',
                                          'neighbourhood_group': 'Zona'})

# máximo precio por noche
df_hood_max = df.groupby(['neighbourhood', 'neighbourhood_group'])['price_night'].max().sort_values(ascending=False).reset_index()
df_hood_max = df_hood_max.rename(columns={'neighbourhood': 'Vecindario', 
                                          'price_night': 'Precio máximo por noche',
                                          'neighbourhood_group': 'Zona'})

# configurando gráfico promedio por noche
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(data=df_hood_avg.head(10), x='Vecindario', y='Precio promedio por noche', hue='Zona', dodge=False)
plt.title('Top 10: Precio promedio por noche por Vecindario')
plt.xlabel('Vecindario')
plt.ylabel('Precio promedio por noche')
plt.xticks(rotation=45)
plt.legend(title='Zona')

# configurando gráfico máximo precio por noche
plt.subplot(1, 2, 2)
sns.barplot(data=df_hood_max.head(10), x='Vecindario', y='Precio máximo por noche', hue='Zona', dodge=False)
plt.title('Top 10: Precio máximo por noche por Vecindario')
plt.xlabel('Vecindario')
plt.ylabel('Precio máximo por noche')
plt.xticks(rotation=45)
plt.legend(title='Zona')

# ajustar diseño y mostrar los gráficos
plt.tight_layout()
plt.show()
```

Podemos evidenciar que Manhattan posee 6 vecindarios en el top 10 de vecindarios con el precio promedio por noche más alto, lo cual no es una sorpresa. También podemos confirmar que los precios por noche más alto son en vecindarios de clase alta, lo cual nos confirma que estos registros podremos tomarlos como atípicos, ya que la ocurrencia de ello no es muy habitual y crean sesgo en nuestros datos.

Procedemos a realizar la distribución del precio promedio por noche.

### Distribución de los precios promedios por noche

Graficaremos la distribución de los precios promedios para verificar nuestra hipótesis de valores atípicos en los datos.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4)) # creamos un lienzo de 1 fila y 2 columnas

sns.boxplot(x=df["price_night"], ax=axes[0]) # creamos el gráfico de caja
axes[0].set_title("Gráfico de Caja") # título del gráfico
axes[0].set_xlabel("Precio(USD)")

sns.histplot(df["price_night"], bins=50, ax=axes[1]) # creamos la distribución
axes[1].set_title("Distribución de precios promedios por noche") # título del gráfico
axes[1].set_xlabel("Precio(USD)") # título del eje x
axes[1].set_ylabel("Frecuencia") # título del eje y

plt.tight_layout() # ajustando gráficos
plt.show() # visualizando
```

Efectivamente nuestros datos estan sesgados hacia la izquierda, necesitamos corregir los valores atípicos ya que estos nos crean sesgos en los datos e interfieren en un correcto análisis de los mismos. Para esto, usaremos el método de rango intercuartílico.

### Método de rango intercuartílico aplicado a price_night

```python
# método de rango intercuartílico
Q1 = df['price_night'].quantile(0.25)
Q3 = df['price_night'].quantile(0.75)

IQR = Q3 - Q1 # Calcular el rango intercuartílico (IQR)

lower_limit = Q1 - 3 * IQR # Definir límites para detectar outliers
upper_limit = Q3 + 3 * IQR

df = df[(df['price_night'] >= lower_limit) & (df['price_night'] <= upper_limit)] # Filtrar valores que están dentro del rango límite
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4)) # creamos un lienzo de 1 fila y 2 columnas

sns.boxplot(x=df["price_night"], ax=axes[0]) # creamos el gráfico de caja
axes[0].set_title("Gráfico de Caja") # título del gráfico
axes[0].set_xlabel("Precio(USD)")

sns.histplot(df["price_night"], bins=50, ax=axes[1]) # creamos la distribución
axes[1].set_title("Distribución de precios promedios por noche") # título del gráfico
axes[1].set_xlabel("Precio(USD)") # título del eje x
axes[1].set_ylabel("Frecuencia") # título del eje y

plt.tight_layout() # ajustando gráficos
plt.show() # visualizando
```

Al remover los datos atípicos usando el Método de rango intercuartílico, se nos visualiza una distribución más acertada y donde podemos ver que el precio mínimo por noche es de 10, el precio máximo por noche es de 270 y la mediana es de 50.

### Distribución de las noches mínimas por estadía

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4)) # creamos un lienzo de 1 fila y 2 columnas

sns.boxplot(x=df["minimum_nights"], ax=axes[0]) # creamos el gráfico de caja
axes[0].set_title("Gráfico de Caja") # título del gráfico
axes[0].set_xlabel("Precio(USD)")

sns.histplot(df["minimum_nights"], bins=50, ax=axes[1]) # creamos la distribución
axes[1].set_title("Distribución de las noches mínimas por estadía") # título del gráfico
axes[1].set_xlabel("Número de noches") # título del eje x
axes[1].set_ylabel("Frecuencia") # título del eje y

plt.tight_layout() # ajustando gráficos
plt.show() # visualizando
```

De igual forma que el precio por noche, podemos notar un sesgo hacia la izquierda de la distribución de las noches mínimas por estadía, esto quiere decir que la gran mayoría de las propiedades manejan la modalidad de estadía por muy pocas noches, esto tra su pro y contra, su pro es que se puede cobrar la estadía por noche un poco más cara y su contra es que la propiedad puede pasar tiempo sin recibir algún inquilino y allí es donde una estadía más prolongada es más beneficiosa para mantener el ingreso.

### 1) Fijar un precio mínimo, máximo y promedio de 25, 1900 y 225 USD respectivamente. Ello está argumentado en el conocimiento de nuestros especialistas en la ciudad

En base al análisis realizado, no estoy de acuerdo en fijar precios. Los precios presentan una dispersión muy alta a consecuencia de la zona donde se encuentra la propiedad, el vecindario y el tipo de propiedad, y por lo tanto, fijar los precios no es conveniente, lo ideal es realizar un análisis aún más detallado y poder promediar los precios en base a la zona, vecindario y tipo de propiedad, esto porque citando un ejemplo, una propiedad puede estar en Manhattan, pero no es lo mismo una propiedad en Time Square a una propiedad en Inwood. Por lo tanto, se deben manejar los precios de las propiedades por cuadrantes.

### 2) Realizar descuentos sobre el precio cuando las estadías son más extensas, al parecer es una práctica del mercado

Con respecto a esta estrategía, si estoy de acuerdo en realizar descuentos de por lo menos un 5% en zonas de clase baja, de 10% en zonas de clase media y un 15% en zonas de clase alta para incentivar la estadía por más días en las propiedades, esto con el fin de mantener un ingreso más estable sobre la propiedad, lo que se traduce en una economía más consistente.

### 3) Unificar tarifas entre Brooklyn y Manhattan

No estoy de acuerdo con esta medida, ya que Manhattan es una zona mucho más exclusiva que Brooklyn y en donde los precios de las propiedades son más altos, esto fundamentado con el análisis que realizamos donde se pudo observar que el precio promedio por noche en Manhattan es mucho mayor que en Brooklyn, por lo tanto, no considero esta estrategia como viable.
