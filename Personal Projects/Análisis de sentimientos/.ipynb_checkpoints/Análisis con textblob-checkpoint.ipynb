# Análisis de sentimientos

Este proyecto busca realizar un análisis de sentimientos usando textblob.

```python
# importando librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier
import warnings

warnings.filterwarnings("ignore")
```

# Importando el dataset

```python
df = pd.read_csv("nyt.csv")
df.head()
```

```python
# La polaridad va entre -1(sentimiento negativo) a 1(sentimiento positivo) donde los valores cercanos a 0 son neutrales
# La polaridad es una medida que indica la positividad o negatividad del sentimiento expresado en un texto
df["polaridad"] = df["content"].apply(lambda x: TextBlob(x).sentiment.polarity)

# la subjetividad va de 0(muy objetivo) a 1(muy subjetivo)
# la objetividad representa un texto puramente objetivo, sin opiniones personales y la subjetividad representa un texto basado en opiniones personales
df["subjetividad"] = df["content"].apply(lambda x: TextBlob(x).sentiment.subjectivity)

df.head()
```

```python
sns.displot(df["polaridad"], kde=True)
sns.displot(df["subjetividad"], kde=True)
plt.show()
```

```python
df.to_csv("Texto analizado.csv")
```
