# Predicción de cáncer de seno

El objetivo de este proyecto es crear un modelo de machine learning usando una red neuronal para predecir el diagnóstico de cáncer de seno en pacientes.

```python
# importandi librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
import warnings

warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
# cargando dataset
df = pd.read_csv('data.csv')
df.head()
```

```python
# información del dataframe
df.info()
```

```python
# verificando valores nulos
df.isnull().sum()
```

```python
# eliminando columnas irrelevantes
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
df.head()
```

```python
# estadisticas del dataframe
df.describe().T
```

```python
# chequeando las clases de la variable objetivo
df['diagnosis'].value_counts()
```

```python
# promedio de las caracteristicas por las clases de la variable objetivo
df.groupby('diagnosis').mean().T
```

```python
# mapeando la variable objetivo
df["diagnosis"] = df["diagnosis"].map({"B":1, "M":0})
df.head()
```

```python
# definiendo las caracteristicas y el objetivo
X = df.drop(columns='diagnosis', axis=1)
y = df['diagnosis']
```

```python
# definiendo datos de prueba y datos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
# estandarizando los datos de prueba y datos de entrenamiento
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# configuracion de la red neuronal usando un modelo secuencial
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)), # capa de aplanamiento de 30 entradas (existen 30 columnas)
                          keras.layers.Dense(20, activation='relu'), # capa densa de 20 neuronas con funcion de activacion relu
                          keras.layers.Dense(2, activation='sigmoid') # capa densa de 2 neuronas con funcion de activacion sigmoid
])
```

```python
# compilando la red neuronal usando el optimizador adam
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', # entropía cruzada categórica dispersa
              metrics=['accuracy']) # métrica de accuracy
```

```python
# entrenando la red neuronal
history = model.fit(X_train, y_train, validation_split=0.1, epochs=10)
```

```python
# visualizando accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
plt.show()
```

```python
# visualizando loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
plt.show()
```

```python
# accuracy y loss del modelo
loss, accuracy = model.evaluate(X_test, y_test)
print(accuracy)
```

```python
# prediccion de los datos de prueba
y_pred = model.predict(X_test)
```

```python
# convirtiendo la probabilidad de predicción en etiquetas de clase
y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```

```python
# sistema de predicción
def prediccion_cancer (data):
    # transformando los datos a un array
    input_data_as_numpy_array = np.asarray(data)

    # reformando el array
    input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

    # estandarizando los datos
    input_data_std = scaler.transform(input_data_reshaped)

    # realizando la prediccion
    prediction = model.predict(input_data_std)

    # convirtiendo la probabilidad de predicción en etiqueta de clase
    prediction_label = [np.argmax(prediction)]
    print(f'La etiqueta resultante es {prediction_label}')

    if(prediction_label[0] == 0):
      print('El tumor es Maligno')
    else:
      print('El tumor es Beningno')
```
