# Predicción de osteoporosis

El objetivo de este proyecto es predecir el riesgo de osteoporosis en pacientes utilizando un conjunto de datos de historiales médicos de pacientes. La osteoporosis es una enfermedad que debilita los huesos, haciéndolos frágiles y más propensos a romperse. Se desarrolla lentamente a lo largo de varios años y a menudo sólo se diagnostica cuando una caída leve o un impacto repentino provocan una fractura ósea. Es más frecuente en las personas mayores, sobre todo.

### Sobre el dataset
El conjunto de datos ofrece información exhaustiva sobre los factores de salud que influyen en el desarrollo de la osteoporosis, incluidos datos demográficos, elecciones de estilo de vida, historial médico e indicadores de salud ósea. Su objetivo es facilitar la investigación en la predicción de la osteoporosis, permitiendo que los modelos de aprendizaje automático identifiquen a las personas en situación de riesgo. El análisis de factores como la edad, el sexo, los cambios hormonales y los hábitos de vida puede ayudar a mejorar la gestión de la osteoporosis y las estrategias de prevención.

### Diccionario del dataset
| Columna | Descripción |
| --- | --- |
|ID| Unique identifier for each patient|
|Age| Age of the patient|
|Gender| Gender of the patient|

|Vitamin D| Vitamin D levels in the patient's body|

|Alcohol Consumption| Whether the patient consumes alcohol|
|Medical Conditions| Medical conditions of the patient|
|Medication| Medication details of the patient|
|Prior Fracture| Whether the patient has had a prior fracture|
|Osteoporosis| Whether the patient has osteoporosis|

### Análisis potencial en este proyecto
- **Modelización predictiva**: Desarrollar modelos de aprendizaje automático para predecir la probabilidad de osteoporosis basándose en las características proporcionadas. Este análisis es crucial para identificar individuos en riesgo de osteoporosis, permitiendo una intervención temprana y estrategias de prevención.

- **Análisis de la importancia de las características**: Determinar la importancia de cada característica en la predicción del riesgo de osteoporosis. Entender qué factores tienen el impacto más significativo en el riesgo de osteoporosis puede proporcionar información sobre los mecanismos subyacentes y guiar las intervenciones específicas.

- **Análisis de correlación**: Examinar las correlaciones entre diferentes características y el riesgo de osteoporosis. La identificación de correlaciones fuertes puede ayudar a identificar posibles factores de riesgo o asociaciones que pueden justificar una mayor investigación o intervención.

- **Análisis de subgrupos**: Analizar cómo varía el riesgo de osteoporosis en diferentes subgrupos en función de factores demográficos, de estilo de vida o de antecedentes médicos. Comprender cómo interactúan los factores de riesgo dentro de los diferentes grupos de población puede informar sobre los enfoques personalizados para la prevención y el tratamiento de la osteoporosis.

- **Interpretación de modelos**: Interpretar los modelos entrenados para entender cómo las diferentes características contribuyen a la predicción del riesgo de osteoporosis. Este análisis puede proporcionar información sobre las relaciones subyacentes entre las variables y ayudar a los profesionales de la salud a tomar decisiones informadas sobre la atención al paciente y las estrategias de gestión.

```python
# importando librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, mean_absolute_error, mean_squared_error, r2_score, root_mean_squared_error
import warnings

warnings.filterwarnings('ignore')
```

```python
# cargando el dataset
df = pd.read_csv('osteoporosis.csv')
df.head()
```

Pocas columnas tienen valores perdidos, por lo que antes de proceder con el análisis, primero trataré los valores perdidos en el dataframe

```python
# columnas con valores faltantes
columns_with_missing_values = df.columns[df.isnull().any()]
# porcentaje de valores faltantes por columnas
print('Porcentaje de valores faltantes por columnas')
for column in columns_with_missing_values:
    print(f'{column}: {round(df[column].isnull().sum() / df.shape[0] * 100, 2)}%')
```

Las columnas Alcohol Consumption y Medications tienen más de un 50% de valores faltantes, sustituiré estos valores perdidos por "None", ya que es posible que el paciente no consuma alcohol ni tome ningún medicamento. Lo mismo ocurre con la columna Medical Conditions.

Sin embargo, las columnas con más del 50% de valores faltantes pueden no ser muy útiles para el análisis, pero aún así las conservo para el 50% restante de los datos.

```python
# reemplazando valores faltantes por None
df.fillna("None",inplace=True)
```

La columna ID es un identificador e irrelevante para el análisis, por lo que eliminaré esta columna.

```python
df = df.drop(columns=['Id'])
```

## Train Test Split

```python
X = df.drop(columns=['Osteoporosis'])
y = df['Osteoporosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

## Modelos a usar para la predicción

Para predecir el riesgo de osteoporosis, utilizaré los siguientes modelos:
- Logistic Regression
- Random Forest Classifier
- Decision Tree Classifier
- Support Vector Classifier

### Logistic Regression

```python
# creando modelo de regresion logistica
logmodel = LogisticRegression()
```

#### Hyperparameter Tuning usando GridSearchCV

```python
# parametros para el grid search
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'penalty': ['l1', 'l2'],
              'solver': ['liblinear'],
              'max_iter': [100, 1000, 2500, 5000],
              'multi_class': ['auto', 'ovr'],
              'random_state': [0,42,101]}

# objeto grid search
grid = GridSearchCV(logmodel,param_grid,refit=True,verbose=3,cv=5,n_jobs=-1)
# entrenando grid search
grid.fit(X_train,y_train)
# imprimiendo mejores parametros
print(grid.best_params_)
```

```python
# parametrizando el modelo
logmodel = LogisticRegression(C=0.1, max_iter=100, penalty='l2', random_state=0, solver='liblinear', multi_class='auto')

# entrenando el modelo
logmodel.fit(X_train,y_train)

# accuracy de las datos de entrenamiento
print(f'Training accuracy: {logmodel.score(X_train,y_train)}')

# prediccion
lr_pred = logmodel.predict(X_test)
```

### Random Forest Classifier

```python
# creando modelo de random forest
rfc = RandomForestClassifier()
```

#### Hyperparameter Tuning usando GridSearchCV

```python
# parametros para el grid search
param_grid = {'criterion': ['gini', 'entropy'],
              'max_depth': [10, 20, 30],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [2,5,10],
              'random_state': [0,42,101]}

# objeto grid search
grid = GridSearchCV(rfc,param_grid,refit=True,verbose=3,cv=5,n_jobs=-1)
# entrenando el grid search
grid.fit(X_train,y_train)
# imprimiendo mejores parametros
print(grid.best_params_)
```

```python
# parametrizando el modelo
rfc = RandomForestClassifier(criterion='gini', max_depth=10, min_samples_leaf=2, min_samples_split=2, random_state=42)

# entrenando el modelo
rfc.fit(X_train,y_train)

# accuracy de los datos de entrenamiento
print(f'Training accuracy: {rfc.score(X_train,y_train)}')

# prediccion
rfc_pred = rfc.predict(X_test)
```

### Decision Tree Classifier

```python
# creando modelo de decision tree
dtree = DecisionTreeClassifier()
```

#### Hyperparameter Tuning usando GridSearchCV

```python
# parametros para el grid search
param_grid = {'criterion': ['gini', 'entropy'],
              'max_depth': [10, 20, 30],
              'min_samples_split': [2, 5, 10],
              'min_samples_leaf': [2,5,10],
              'random_state': [0,42,101]}

# objeto grid search
grid = GridSearchCV(dtree,param_grid,refit=True,verbose=3,cv=5,n_jobs=-1)
# entrenando el grid search
grid.fit(X_train,y_train)
# imprimiendo mejores parametros
print(grid.best_params_)
```

```python
# parametrizando el modelo
dtree = DecisionTreeClassifier(criterion='entropy', max_depth=10, min_samples_leaf=10, min_samples_split=2, random_state=0)

# entrenando el modelo
dtree.fit(X_train,y_train)

# accuracy de los datos de entrenamiento
print(f'Training accuracy: {dtree.score(X_train,y_train)}')

# prediccion
dtree_pred = dtree.predict(X_test)
```

### Support Vector Classifier

```python
# creando modelo SVM
svc = SVC()
```

#### Hyperparameter Tuning usando GridSearchCV

```python
# parametros para el grid search
param_grid = {'C': [0.1, 1, 10, 100],
              'degree': [2, 3, 4, 5],
              'gamma': ['scale', 'auto'],
              'random_state': [0,42,101]}

# objeto grid search
grid = GridSearchCV(svc,param_grid,refit=True,verbose=3,cv=5,n_jobs=-1)
# entrenando el grid search
grid.fit(X_train,y_train)
# imprimiendo mejores parametros
print(grid.best_params_)
```

```python
# parametrizando el modelo
svc = SVC(C=0.1, degree=2, gamma='auto', random_state=0, kernel='linear')

# entrenando el modelo
svc.fit(X_train,y_train)

# accuracy de los datos de entrenamiento
print(f'Training accuracy: {svc.score(X_train,y_train)}')

# prediccion
svc_pred = svc.predict(X_test)
```

## Evaluación de los modelos

### Matrix de confusión

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 15))

#confusion matrix for logistic regression
cm = confusion_matrix(y_test, lr_pred)
sns.heatmap(cm, annot=True, ax = ax[0,0], fmt='g').set_title('Logistic Regression')

#confusion matrix for random forest
cm = confusion_matrix(y_test, rfc_pred)
sns.heatmap(cm, annot=True, ax = ax[0,1], fmt='g').set_title('Random Forest Classifier')

#confusion matrix for decision tree
cm = confusion_matrix(y_test, dtree_pred)
sns.heatmap(cm, annot=True, ax = ax[1,0], fmt='g').set_title('Decision Tree Classifier')

#confusion matrix for support vector classifier
cm = confusion_matrix(y_test, svc_pred)
sns.heatmap(cm, annot=True, ax = ax[1,1], fmt='g').set_title('Support Vector Classifier')

plt.show()
```

### Accuracy de los modelos

```python
# grafico de barra para el accuracyde los modelos
models = ['Logistic Regression', 'Random Forest', 'Decision Tree', 'Support Vector Classifier']
accuracy = [accuracy_score(y_test, lr_pred), accuracy_score(y_test, rfc_pred), accuracy_score(y_test, dtree_pred), accuracy_score(y_test, svc_pred)]
plt.figure(figsize=(10,5))
sns.barplot(x=models, y=accuracy).set_title('Accuracy de los modelos')
plt.show()
```

### Metricas de los modelos

```python
fig, ax = plt.subplots(2,2, figsize=(15, 15))
models = ['Logistic Regression', 'Random Forest', 'Decision Tree', 'Support Vector Classifier']
mae = [mean_absolute_error(y_test, lr_pred), mean_absolute_error(y_test, rfc_pred), mean_absolute_error(y_test, dtree_pred), mean_absolute_error(y_test, svc_pred)]
mse = [mean_squared_error(y_test, lr_pred), mean_squared_error(y_test, rfc_pred), mean_squared_error(y_test, dtree_pred), mean_squared_error(y_test, svc_pred)]
rmse = [np.sqrt(mean_squared_error(y_test, lr_pred)), np.sqrt(mean_squared_error(y_test, rfc_pred)), np.sqrt(mean_squared_error(y_test, dtree_pred)), np.sqrt(mean_squared_error(y_test, svc_pred))]
r2 = [r2_score(y_test, lr_pred), r2_score(y_test, rfc_pred), r2_score(y_test, dtree_pred), r2_score(y_test, svc_pred)]

# Mean Absolute Error
sns.barplot(x=models, y=mae, ax=ax[0,0]).set_title('Mean Absolute Error')

# Mean Squared Error
sns.barplot(x=models, y=mse, ax=ax[0,1]).set_title('Mean Squared Error')

# Root Mean Squared Error
sns.barplot(x=models, y=rmse, ax=ax[1,0]).set_title('Root Mean Squared Error')

# R2 Score
sns.barplot(x=models, y=r2, ax=ax[1,1]).set_title('R2 Score')

plt.show()
```

## Feature Importance Analysis

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 20))

# Feature Importance graph for Logistic Regression
coeff = list(logmodel.coef_[0])
labels = list(df.drop('Osteoporosis',axis=1).columns)
features = pd.DataFrame()
features['Features'] = labels
features['Importance'] = coeff
features.sort_values(by=['Importance'], ascending=True, inplace=True)
features = features.set_index('Features')
features.plot(kind='barh', ax=ax[0,0]).set_title('Feature Importance for Logistic Regression')

# Feature Importance graph for Random Forest
coeff = list(rfc.feature_importances_)
labels = list(df.drop('Osteoporosis',axis=1).columns)
features = pd.DataFrame()
features['Features'] = labels
features['Importance'] = coeff
features.sort_values(by=['Importance'], ascending=True, inplace=True)
features = features.set_index('Features')
features.plot(kind='barh', ax=ax[0,1]).set_title('Feature Importance for Random Forest')

# Feature Importance graph for Decision Tree
coeff = list(dtree.feature_importances_)
labels = list(df.drop('Osteoporosis',axis=1).columns)
features = pd.DataFrame()
features['Features'] = labels
features['Importance'] = coeff
features.sort_values(by=['Importance'], ascending=True, inplace=True)
features = features.set_index('Features')
features.plot(kind='barh', ax=ax[1,0]).set_title('Feature Importance for Decision Tree')

# Feature Importance graph for Support Vector Classifier
coeff = list(svc.coef_[0])
labels = list(df.drop('Osteoporosis',axis=1).columns)
features = pd.DataFrame()
features['Features'] = labels
features['Importance'] = coeff
features.sort_values(by=['Importance'], ascending=True, inplace=True)
features = features.set_index('Features')
features.plot(kind='barh', ax=ax[1,1]).set_title('Feature Importance for Support Vector Classifier')

plt.show()
```

## Conclusión

En este proyecto, desarrollé modelos de aprendizaje automático para predecir el riesgo de osteoporosis en pacientes basándome en sus historiales médicos. Analicé el conjunto de datos, realicé un análisis exploratorio de datos y desarrollé modelos predictivos utilizando regresión logística, clasificador de bosque aleatorio, clasificador de árbol de decisión y clasificador de vector de soporte. Los modelos se evaluaron mediante las métricas de matriz de confusión, exactitud, precisión, recuperación y puntuación F1.

A partir del análisis exploratorio de datos, he descubierto que ciertos factores como la edad, los cambios hormonales, las condiciones médicas, la medicación, el estilo de vida y la nutrición son responsables del riesgo de osteoporosis. Los pacientes de entre 20 y 40 años tienen menor riesgo de osteoporosis. Los pacientes que han sufrido cambios hormonales, tienen afecciones médicas, consumen medicamentos, tienen menor peso corporal, niveles de calcio y vitamina D, y llevan un estilo de vida sedentario tienen mayor riesgo de osteoporosis.

En cuanto a los modelos de aprendizaje automático, he empleado regresión logística, árbol aleatorio, árbol de decisión y clasificador de vectores de apoyo para predecir el riesgo de osteoporosis basándome en los datos. De estos modelos, el clasificador de árbol de decisión fue el que dio mejores resultados en comparación con los demás, con casi un 87% de precisión. El modelo puede utilizarse para predecir el riesgo de osteoporosis en pacientes basándose en sus historiales médicos, lo que permite una intervención temprana y estrategias de prevención.
