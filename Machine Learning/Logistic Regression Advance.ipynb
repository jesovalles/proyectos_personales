```python
X = df.drop('booking_status', axis = 1)
Y = df['booking_status']
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state=1)
```

```python
def metricas(real, predecido):
    print(classification_report(real, predecido))
    
    cm = confusion_matrix(real, predecido)
    plt.figure(figsize=(8,5))
    sns.heatmap(cm, annot=True, fmt='.2f', xticklabels = ['No Cancelada', 'Cancelada'], yticklabels = ['No Cancelada', 'Cancelada'])
    
    plt.ylabel('Real')
    plt.xlabel('Predecido')
    
    plt.show()
```

```python
reg_log = LogisticRegression(random_state = 1)

reg_log.fit(X_train, y_train)
```

```python
pd.Series(reg_log.coef_[0], index = X_train.columns).sort_values(ascending=False)
```

```python
y_pred_train = reg_log.predict(X_train)

metricas(y_train, y_pred_train)
```

```python
y_scores = reg_log.predict_proba(X_train) 

precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores[:,1])

# Plotting values of precisions, recalls, and thresholds
plt.figure(figsize = (10, 7))

plt.plot(thresholds, precisions[:-1], 'b--', label = 'precision')

plt.plot(thresholds, recalls[:-1], 'g--', label = 'recall')

plt.xlabel('Threshold')

plt.legend(loc = 'upper left')

plt.ylim([0, 1])

plt.show()
```

```python
for i in np.arange(len(thresholds)):
    if precisions[i] == recalls[i]:
        print(thresholds[i])
```

```python
optimo = 0.42

metricas(y_train, y_scores[:, 1] > optimo)
```

```python
y_pred_test = reg_log.predict_proba(X_test)

metricas(y_test, y_pred_test[:, 1] > optimo)
```
