```python
# importar librerias
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
#Lectura del archivo a analizar
df = pd.read_csv('train.csv')
df.describe()
```

```python
#GrÃ¡fica de dispersiÃ³n
plt.scatter(df["GrLivArea"], df["SalePrice"], alpha=0.5)
plt.xlabel("Metros cuadrados")
plt.ylabel("Precio de la casa")
plt.show()
```

```python
#CreaciÃ³n de modelo de regresiÃ³n lineal simple y= mx + b
regresion = linear_model.LinearRegression()

#Convertir los datos de la columna GrLivArea a un vector que serÃ¡ la variable independiente
df_GrLivArea = df["GrLivArea"].values.reshape((-1,1))

#Crear modelo en base a la variable independiente para hallar la variable dependiente
modelo = regresion.fit(df_GrLivArea, df["SalePrice"])

# valor de la pendiente
m = modelo.coef_[0]

# valor de la intercepciÃ³n con el eje Y
b = modelo.intercept_

print(f"Pendiente (m) = {m:.2f}")
print(f"IntercepciÃ³n con el eje Y (b) = {b:.2f}")
print(f"- La pendiente indica que por cada metro cuadrado, el precio aumenta = {m:.2f}")
```

```python
# Predicciones del modelo
predicciones = modelo.predict(df_GrLivArea)

# CÃ¡lculo de errores
mse = mean_squared_error(df["SalePrice"], predicciones)
r2 = r2_score(df["SalePrice"], predicciones)
mae = mean_absolute_error(df["SalePrice"], predicciones)

print(f"Error cuadrÃ¡tico medio (MSE) = {mse:.2f}")
print(f"Coeficiente de determinaciÃ³n (R2) = {r2:.2f}")
print(f"Error Absoluto Medio (MAE) = {mae:.2f}")
```

```python
#GrÃ¡fica de dispersiÃ³n + modelo de regresiÃ³n lineal simple
plt.scatter(df["GrLivArea"], df["SalePrice"], color="blue", label="Datos", alpha=0.1)
plt.plot(df["GrLivArea"], modelo.predict(df[["GrLivArea"]]), color="black", label="Modelo de regresiÃ³n lineal")
plt.xlabel("Metros cuadrados")
plt.ylabel("Precio de la casa")
plt.legend()
plt.show()
```

```python
# funcion para predecir precio de las casas en base a los metros cuadrados
def prediction_function(ft):
    y = round((m * ft + b), 2)
    return y
```
