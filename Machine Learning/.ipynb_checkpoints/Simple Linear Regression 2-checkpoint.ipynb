```python
# importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
#Lectura del archivo a analizar
df = pd.read_csv('train.csv')
df.describe()
```

```python
#GrÃ¡fica de dispersiÃ³n
plt.scatter(df["GrLivArea"], df["SalePrice"], alpha=0.5)
plt.xlabel("GrLivArea")
plt.ylabel("SalePrice")
plt.show()
```

```python
# Reshape de los datos para que tengan la forma correcta
x = df["GrLivArea"].values.reshape((-1,1))
y = df["SalePrice"].values.reshape((-1,1))

# Dividir los datos en conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Crear el modelo de regresiÃ³n lineal
model = linear_model.LinearRegression()

# Entrenar el modelo con los datos de entrenamiento
model.fit(x_train, y_train)
```

```python
# Coeficientes del modelo
m = model.coef_[0]
b = model.intercept_

print(f"Pendiente (m) = {m[0]:.2f}")
print(f"IntercepciÃ³n con el eje Y (b) = {b[0]:.2f}")
print(f"- La pendiente indica que por cada metro cuadrado, el precio aumenta = {m[0]:.2f}")
```

```python
# Realizar predicciones en los datos de prueba
y_pred = model.predict(x_test)

# Calcular los errores
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(f"Error cuadrÃ¡tico medio (MSE) = {mse:.2f}")
print(f"Coeficiente de determinaciÃ³n (R2) = {r2:.2f}")
print(f"Error Absoluto Medio (MAE) = {mae:.2f}")
```

```python
# VisualizaciÃ³n de los resultados
plt.scatter(x_train, y_train, label='Datos de entrenamiento')
plt.scatter(x_test, y_test, color='red', label='Datos de prueba')
plt.plot(x_test, y_pred, color='gold', label='RegresiÃ³n lineal')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.legend()
plt.show()

print(f"Coeficiente (pendiente): {m[0]:.2f}")
print(f"Interceptor: {b[0]:.2f}")
print(f"Error cuadrÃ¡tico medio (MSE): {mse:.2f}")
print(f"Coeficiente de determinaciÃ³n (R^2): {r2:.2f}")
print(f"Error absoluto medio (MAE) (MAE): {mae:.2f}")
```

```python
# funcion para predecir precio de las casas en base a los metros cuadrados
def prediction_function(ft):
    y = round((m[0] * ft + b[0]), 2)
    return y
```
