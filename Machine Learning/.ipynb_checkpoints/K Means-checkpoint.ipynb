#### K Means Clustering - Customer Segmentation

```python
# importando librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
# cargando la data del csv a un dataframe llamado df
df = pd.read_csv('Mall customers.csv')
```

```python
# chequeando primeras 5 filas del dataframe
df.head()
```

```python
# verificando el numero de filas y columnas del dataframe
df.shape
```

```python
# obteniendo informacion sobre el dataframe
df.info()
```

```python
# revisando si las columnas del dataframe contienen valores nulos
df.isnull().sum()
```

```python
# graficando los datos a analizar
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Graficando los datos a analizar por genero
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', data=df)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# escogiendo todas las filas de las columnas Annual Income Column & Spending Score column
X = df.iloc[:,[3,4]].values
```

```python
# escogiendo el numero de clusters
# WCSS  ->  Within Clusters Sum of Squares
# encontrando el valor WCSS para cada diferentes numeros de clusters

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++')
  kmeans.fit(X)

  wcss.append(kmeans.inertia_)
```

```python
# graficando el elbow graph o grafico de codo para determinar el numero de clusters a usar optimo
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Al observar la gráfica, el número óptimo de clusters para este modelo es de 5

```python
# entrenando al modelo K-Means Clustering
kmeans = KMeans(n_clusters=5, init='k-means++')

# asigna cada valor a uno de los clusters
Y = kmeans.fit_predict(X)
```

```python
# graficando cada clusters (plt.figure(figsize=(8,6)))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster 5')

# graficando los centroides
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='Centroides')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# graficando cada clusters (plt.figure(figsize=(8,6)))
plt.scatter(X[Y==0,0], X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=50, c='blue', label='Cluster 5')

# graficando los centroides
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='Centroides')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Función para asignar el valor a un cluster
def clasificacion(x):
    cluster = kmeans.predict([x])
    return f'El nuevo punto pertenece al Cluster {cluster[0] + 1}'

punto = [55, 50]  #[Annual Income, Spending Score]
resultado = clasificacion(punto)
print(resultado)
```
